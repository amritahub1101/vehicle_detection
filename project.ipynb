```python
import ultralytics as ut
from ultralytics import YOLO
import cv2

model = YOLO("yolo11n.pt")
image = "road.jpg"
results = model(image, conf=0.25)

output = results[0].plot()
counts = {}

# Count objects
for box in results[0].boxes:
    class_id = int(box.cls[0])
    object_name = model.names[class_id]
    if object_name in counts:
        counts[object_name] += 1
    else:
        counts[object_name] = 1

print("\n------ Detection Summary ------")
total = 0

for name, count in counts.items():
    print(f"{name} : {count}")
    total += count

print("-------------------------------")
print("Total Objects Detected:", total)

# Display counts on image
y = 30
cv2.putText(output, f"Total Objects: {total}", (10, y),
             cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
y += 35

for name, count in counts.items():
    cv2.putText(output, f"{name}: {count}", (10, y),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
    y += 30
cv2.imwrite("detected_output.jpg", output)
cv2.imshow("Object Detection and Counting", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
